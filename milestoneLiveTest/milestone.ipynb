```python
# Q1 what is the difference between static and dynamic variable variables in python?
'''
 @static variable:-

 Static variable is also known global variable.
 It defines within the class but  outside the methods
 they shared across all instance of the class
 These variables are used when you want to maintain a common value for all instances of a class.
 '''
# static variable example
class MyClass:
    static_var = 42  # This is a static variable

obj1 = MyClass()
obj2 = MyClass()

print(obj1.static_var)  # Output: 42
print(obj2.static_var)  # Output: 42

MyClass.static_var = 100
print(obj1.static_var)  # Output: 100
print(obj2.static_var)  # Output: 100
'''
 @dynamic variable:-

 Dynamic variable are define within the method and are unique to each instance of class.
 These variables are used when each instance of a class needs to maintain its own state.

 '''
# dynamic variable example:-
class MyClass:
    def __init__(self, value):
        self.dynamic_var = value  # This is a dynamic variable

obj1 = MyClass(10)
obj2 = MyClass(20)

print(obj1.dynamic_var)  # Output: 10
print(obj2.dynamic_var)  # Output: 20

```

```python
# Q18 what is itrators ,itrable and generators in python/?

# Iterable
# An iterable is any Python object capable of returning its members one at a time. Examples of iterables include lists, tuples, strings, dictionaries, and sets. Iterables implement the __iter__ method, which returns an iterator.

my_tuple = (1, 2, 3)
for item in my_tuple:
    print(item)
my_list = [1, 2, 3]
for item in my_list:
    print(item)

# Iterator
# An iterator is an object representing a stream of data; it is an object that can be iterated upon (i.e., you can traverse through all the values). Iterators implement the __iter__ and __next__ methods.
# **The __iter__ method returns the iterator object itself.
# **The __next__ method returns the next value from the iterator. If there are no more items to return, it raises a StopIteration exception.

# Creating an iterator from a list
my_list = [1, 2, 3]
iterator = iter(my_list)

# Using the iterator
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3

# This will raise StopIteration
# print(next(iterator))

# generators
# In Python, generators are a special type of iterator that allow you to iterate over data without needing to store the entire dataset in memory. They are defined using functions and the yield keyword.
def count_up_to(max_value):
    counter = 1
    while counter <= max_value:
        yield counter
        counter += 1

# Using the generator
counter_gen = count_up_to(5)
for num in counter_gen:
    print(num)  # Output: 1 2 3 4 5

```

```python
# Q22 How does inheritance works in python? Explain all types of inheritance with an example.

'''Inheritance is a fundamental concept in object-oriented programming that allows a class to inherit attributes and methods from another class. This promotes code reusability and establishes a natural hierarchy between classes.

In Python, there are several types of inheritance:

Single Inheritance
Multiple Inheritance
Multilevel Inheritance
Hierarchical Inheritance
Hybrid Inheritance'''

# 1. Single inheritance
class Parent:
    def display(self):
        print("This is the Parent class.")

class Child(Parent):
    pass

child = Child()
child.display()  # Output: This is the Parent class.


# 2. Multiple Inheritance
class Parent1:
    def display1(self):
        print("This is Parent1 class.")

class Parent2:
    def display2(self):
        print("This is Parent2 class.")

class Child(Parent1, Parent2):
    pass

child = Child()
child.display1()  # Output: This is Parent1 class.
child.display2()  # Output: This is Parent2 class.


# 3. Multilevel Inheritance
class Grandparent:
    def display_grandparent(self):
        print("This is the Grandparent class.")

class Parent(Grandparent):
    def display_parent(self):
        print("This is the Parent class.")

class Child(Parent):
    pass

child = Child()
child.display_grandparent()  # Output: This is the Grandparent class.
child.display_parent()       # Output: This is the Parent class.


# 4. Hierarchical Inheritance
class Parent:
    def display(self):
        print("This is the Parent class.")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

child1 = Child1()
child2 = Child2()
child1.display()  # Output: This is the Parent class.
child2.display()  # Output: This is the Parent class.

```

```python
# Q23 what is encapsulation? explain it with an example.

'''
Encapsulation is a fundamental concept in object-oriented programming that refers to the bundling of data (attributes) and methods (functions) that operate on the data into a single unit or class. It also involves restricting direct access to some of an object's components, which can help prevent the accidental modification of data. This is typically achieved using access modifiers.

Key Points of Encapsulation:
Private Attributes/Methods: These are not accessible from outside the class. In Python, this is achieved by prefixing the attribute or method name with double underscores (__).
Public Attributes/Methods: These are accessible from outside the class.
Protected Attributes/Methods: These are indicated by a single underscore (_) and are intended to be used within the class and its subclasses.'''

class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number  # Public attribute
        self.__balance = balance              # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Invalid withdrawal amount.")

    def get_balance(self):
        return self.__balance

# Creating an instance of BankAccount
account = BankAccount("123456", 1000)

# Accessing public attribute
print("Account Number:", account.account_number)  # Output: Account Number: 123456

# Trying to access private attribute (will raise an AttributeError)
try:
    print("Balance:", account.__balance)
except AttributeError as e:
    print(e)  # Output: 'BankAccount' object has no attribute '__balance'

# Using public methods to interact with private attribute
account.deposit(500)
print("Balance after deposit:", account.get_balance())  # Output: Balance after deposit: 1500

account.withdraw(200)
print("Balance after withdrawal:", account.get_balance())  # Output: Balance after withdrawal: 1300

# Attempting to access the private attribute directly using name mangling
print("Balance (using name mangling):", account._BankAccount__balance)  # Output: Balance (using name mangling): 1300



```

```python
# 1.6. Write a program to display the appropriate message as per the color of signal(RED-Stop/Yellow-Stay/Green-Go) at the road crossing. 

def traffic_signal_message(color):
    if color.lower() == "red":
        return "Stop"
    elif color.lower() == "yellow":
        return "Stay"
    elif color.lower() == "green":
        return "Go"
    else:
        return "Invalid color"

# Main program
while True:
    color = input("Enter the traffic signal color (RED, YELLOW, GREEN) or 'exit' to quit: ").strip()
    
    if color.lower() == "exit":
        print("Exiting the program.")
        break
    
    message = traffic_signal_message(color)
    print(message)
```

```python
# 1.7. Write a program to create a simple calculator performing only four basic operations(+,-,/,*) .

# calculator
# Define functions for each operation
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

# Display the menu to the user
def menu():
    print("Simple Calculator")
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

while True:
    menu()
    
    # Take input from the user
    choice = input("Enter choice (1/2/3/4/5): ")

    # Check if choice is one of the four options or exit
    if choice in ['1', '2', '3', '4']:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print(f"{num1} + {num2} = {add(num1, num2)}")

        elif choice == '2':
            print(f"{num1} - {num2} = {subtract(num1, num2)}")

        elif choice == '3':
            print(f"{num1} * {num2} = {multiply(num1, num2)}")

        elif choice == '4':
            print(f"{num1} / {num2} = {divide(num1, num2)}")

    elif choice == '5':
        print("Exiting the calculator.")
        break

    else:
        print("Invalid input. Please enter a number between 1 and 5.")


```

```python
# 1.9.  Write a program to find the factors of a whole number  using a while loop.

def factor(num):
    factors = []
    i = 1
    while i<=num:
        if num%i == 0:
            factors.append(i)
        i+=1

    return factors

number = int(input("Enter a whole number: "))

if number <= 0:
    print("Please enter a positive whole number.")
else:
    factors = factor(number)
    print(f"The factors of {number} are: {factors}")


```

```python
#  1.11.   Write a program to find prime numbers between 2 to 100 using nested for loops.

# Function to find and print prime numbers between 2 and 100
def find_primes():
    for num in range(2, 101):  # Loop through numbers from 2 to 100
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):  # Check for factors up to the square root of num
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=' ')  # Print the prime number

print("Prime numbers between 2 and 100 are:")
find_primes()


```

```python
# 4.Create a simple banking system using object-oriented concepts in Python. Design classes representing 
# different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw, 
# and balance inquiry. Utilize inheritance to manage different account types efficiently.

#  Tasks
#  1. Define base class(es) for bank accounts with common attributes and methods
#  2. Implement subclasses for specific account types (e.g., SavingsAccount, CheckingAccount)
#  3. Provide methods for deposit, withdraw, and balance inquiry in each subclass
#  4. Test the banking system by creating instances of different account types and performing transactions


class BankAccount :
    def __init__(self,account_number,account_holder) -> None:
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = 0.0

    def deposite(self,amount):
        if amount>0:
            self.balance+=amount
        else:
            print("amount must be positive")
    def withdraw(self,amount):
        if amount>0: 
            if self.balance>=amount:
                self.balance-=amount
                print(f"withdrew {amount}. New balance is {self.balance}")
            else:
                print("insufficient funds.")
        else:
           print("amount must be positive.")


    def get_balance(self):
        return self.balance

class SavingAccount(BankAccount):
    def __init__(self, account_number, account_holder,interest_rate) -> None:
        super().__init__(account_number, account_holder)
        self.interest_rate = interest_rate

    def apply_interest(self):
        interest = self.balance * self.interest_rate/100
        self.balance += interest
        print(f"applied interest {interest}. New balance {self.balance}")
        
class ChackingAccount(BankAccount):
    def __init__(self, account_number, account_holder,overdraft_limit) -> None:
        super().__init__(account_number, account_holder)
        self.overdraft_limit = overdraft_limit

    def withdrew(self,amount):
        if amount > 0:
            if self.balance + self.overdraft_limit >= amount:
                self.balance -= amount
                print(f"Withdrew ${amount}. New balance is ${self.balance}.")
            else:
                print("Overdraft limit exceeded.")
        else:
            print("Withdrawal amount must be positive.")  

if __name__ == "__main__":
    savings = SavingAccount(1264,"alic",6.2)
    print(savings.get_balance())
    savings.deposite(500)
    savings.apply_interest()

    checking = ChackingAccount(4526,"abhay",500)
    checking.deposite(500)
    checking.withdraw(100)
    checking.withdraw(600)  # This should work due to overdraft limit
    checking.withdraw(100) 
```

```python
# 23. What is correlation. Explain its type in details.what are the  methods of determining correlation
"""Correlation: Definition and Explanation
Definition:
Correlation is a statistical measure that describes the strength and direction of a relationship between two variables. It indicates how changes in one variable are associated with changes in another variable.

Types of Correlation
    Positive Correlation:

        When two variables move in the same direction, they are positively correlated.
        As one variable increases, the other variable also increases, and vice versa.
        Example: Height and weight of individuals.
    
    Negative Correlation:

        When two variables move in opposite directions, they are negatively correlated.
        As one variable increases, the other variable decreases, and vice versa.
        Example: The speed of a car and the travel time to a destination.
    
    Zero Correlation:

        When there is no relationship between two variables, they have zero correlation.
        Changes in one variable do not affect the other variable.
        Example: The number of books read and the number of apples eaten.

Methods of Determining Correlation
    Pearson Correlation Coefficient:

        Measures the linear relationship between two continuous variables.
        Values range from -1 to +1.
            +1 indicates a perfect positive linear relationship.
            -1 indicates a perfect negative linear relationship.
            0 indicates no linear relationship.
        
    Spearman's Rank Correlation Coefficient:

        Measures the strength and direction of the relationship between two ranked variables.
        Non-parametric method (does not assume a normal distribution).
        Values range from -1 to +1, similar to Pearson's.
   
        """

import numpy as np

# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Pearson correlation coefficient
pearson_corr = np.corrcoef(x, y)[0, 1]
print("Pearson Correlation Coefficient:", pearson_corr)


from scipy.stats import spearmanr

# Example data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Spearman's rank correlation coefficient
spearman_corr, _ = spearmanr(x, y)
print("Spearman's Rank Correlation Coefficient:", spearman_corr)



```

```python
# 28. What is  Normal Distribution? What are the four Assumptions of Normal Distribution? Explain in detail.
"""
Normal Distribution

Definition:

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetrical around its mean. It is characterized by its bell-shaped curve, which is defined by the probability density function (PDF):

    Properties:

        Symmetry: The distribution is perfectly symmetrical around the mean.
        
        Mean, Median, Mode: These three measures of central tendency are all equal in a normal distribution.
        
        Bell Shape: The shape of the curve is bell-shaped, with tails that asymptotically approach the x-axis.
        Empirical Rule (68-95-99.7 Rule):

        
Assumptions of Normal Distribution
    Independence:

        Definition: The observations are independent of each other.
        
        Explanation: Each data point in the dataset is collected independently, meaning the occurrence of one data point does not influence the occurrence of another.
        
        Example: If you're measuring the heights of individuals, the height of one person should not affect the height of another.
    Random Sampling:

        Definition: The data is collected through a random sampling method.
        
        Explanation: Every member of the population has an equal chance of being included in the sample.
        
        Example: Selecting students randomly from a large university to measure their test scores ensures that the sample represents the population.
    
    Continuous Data:

        Definition: The data should be continuous and not discrete.
        Explanation: Normal distribution applies to continuous data, where measurements can take any value within a range.
        Example: Heights, weights, and temperatures are continuous variables, while the number of children in a family is a discrete variable.
        
    Central Limit Theorem (for large samples):

        Definition: The sample means will be approximately normally distributed, regardless of the shape of the population distribution, provided the sample size is sufficiently large (usually n ≥ 30).
        
        Explanation: Even if the underlying data is not normally distributed, the distribution of the sample means will tend to be normal.
        
        Example: Rolling a die multiple times and calculating the average result will produce a distribution of averages that approximates normal distribution as the number of rolls increases.



"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Create a range of x values
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)

# Calculate the PDF of the normal distribution
pdf = norm.pdf(x, mu, sigma)

# Plot the normal distribution curve
plt.plot(x, pdf, label='Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.axvline(mu, color='r', linestyle='--', label='Mean')
plt.legend()
plt.show()

```

```python
# 29.Write all the characteristics or Properties of the Normal Distribution Curve.


"""
Characteristics or Properties of the Normal Distribution Curve
    Symmetry:

        The normal distribution curve is perfectly symmetrical around the mean (μ). This means that the left half of the curve is a mirror image of the right half.
        
    Mean, Median, and Mode: 
        
        In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution. This central point is the highest point of the curve.

    Bell-Shaped Curve:

        The curve of the normal distribution is bell-shaped and smooth, extending infinitely in both directions along the horizontal axis but never touching it.

    Asymptotic:

        The tails of the normal distribution curve approach, but never touch, the horizontal axis. This implies that the probability of the random variable taking on any value extends to positive and negative infinity, but the probabilities of extreme values are very low.
    
    Empirical Rule (68-95-99.7 Rule):

        Approximately 68% of the data falls within one standard deviation (μ±σ) of the mean.
        
        Approximately 95% falls within two standard deviations (μ±2σ) of the mean.
        
        Approximately 99.7% falls within three standard deviations (μ±3σ) of the mean.
    
    Unimodal:

        The normal distribution has a single peak, meaning it is unimodal. This peak is at the mean of the distribution.
    
    Defined by Mean and Standard Deviation:

        The shape and location of the normal distribution curve are determined by its mean (μ) and standard deviation (σ). The mean determines the center of the distribution, while the standard deviation determines the spread or width of the curve.
    
    Total Area Under the Curve:

        The total area under the normal distribution curve is equal to 1. This represents the total probability of all possible outcomes for a normal random variable.
    
    No Skewness:

        A normal distribution has no skewness. Skewness is a measure of asymmetry of the probability distribution of a real-valued random variable. Since the normal distribution is symmetric, its skewness is 0.


"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Create a range of x values
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)

# Calculate the PDF of the normal distribution
pdf = norm.pdf(x, mu, sigma)

# Plot the normal distribution curve
plt.plot(x, pdf, label='Normal Distribution')
plt.axvline(mu, color='r', linestyle='--', label='Mean')
plt.fill_between(x, pdf, alpha=0.1)
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.legend()
plt.show()
```

```python
# Explain the bias-variance tradeoff
"""
The bias-variance tradeoff is a fundamental concept in machine learning that describes the tradeoff between two sources of error that affect the performance of predictive models: bias and variance. Understanding and managing this tradeoff is crucial for building models that generalize well to new, unseen data.

Bias:
    Definition: Bias refers to the error introduced by approximating a real-world problem, which may be complex, by a simplified model.
    High Bias: Indicates a model is too simple, underfitting the data, and unable to capture the underlying patterns. This leads to systematic errors in predictions.
    Example: Using a linear model to fit a complex, nonlinear relationship.

Variance:
    Definition: Variance refers to the error introduced by the model's sensitivity to small fluctuations in the training data. High variance can cause the model to capture noise in the training data rather than the underlying pattern.
    High Variance: Indicates a model is too complex, overfitting the data, and performing well on the training data but poorly on unseen data.
    Example: Using a high-degree polynomial model to fit a simple linear relationship.

The Tradeoff:
    Low Bias and High Variance: The model is very flexible and fits the training data very well, but it may not generalize to new data (overfitting).
    High Bias and Low Variance: The model is too rigid and does not fit the training data well, leading to poor performance on both the training and new data (underfitting).   

"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate synthetic data
np.random.seed(0)
X = np.sort(np.random.rand(30) * 10)
y = np.sin(X) + np.random.normal(0, 0.1, X.shape)

# Plot original data
plt.scatter(X, y, label='Data', color='black')

# Fit models with varying complexity
degrees = [1, 4, 15]
for degree in degrees:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X[:, np.newaxis], y)
    y_pred = model.predict(X[:, np.newaxis])
    
    plt.plot(X, y_pred, label=f'Degree {degree}')

plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Bias-Variance Tradeoff')
plt.show()



```

```python
#  What are precision and recall? How are they different from accuracy
"""
Precision, recall, and accuracy are three key metrics used to evaluate the performance of classification models in machine learning. Each metric provides different insights into the model's performance, especially in the context of imbalanced datasets. Here's a detailed explanation of each:

Precision:
    Definition: Precision is the ratio of true positive predictions to the total number of positive predictions (both true positives and false positives). It answers the question: "Of all the instances predicted as positive, how many are actually positive?"

    Formula:
        Precision = True Positives/(True Positives+False Positives)

    Interpretation: High precision indicates that the model has a low false positive rate, meaning it rarely misclassifies negative instances as positive.

Recall:
    Definition: Recall (also known as Sensitivity or True Positive Rate) is the ratio of true positive predictions to the total number of actual positive instances (both true positives and false negatives). It answers the question: "Of all the actual positive instances, how many did the model correctly identify?"

    Formula:
        Recall = True Positives/( True Positives + False Negatives)

    Interpretation: High recall indicates that the model has a low false negative rate, meaning it captures most of the actual positive instances.

Accuracy:
    Definition: Accuracy is the ratio of correct predictions (both true positives and true negatives) to the total number of predictions. It answers the question: "How often is the model correct?"

    Formula:
        Accuracy = True Positives + True Negatives/ Total Predictions

    Interpretation: Accuracy gives a general idea of how well the model performs, but it can be misleading, especially with imbalanced datasets where one class is much more frequent than the other.


Precision/Recall vs. Accuracy:

    Accuracy gives an overall measure of performance but can be misleading if classes are imbalanced. For example, if 90% of instances belong to the negative class, a model that always predicts negative will have high accuracy but poor precision and recall for the positive class.
    
    Precision and Recall provide more detailed insights into the performance on positive predictions, which is especially important in imbalanced datasets or in applications where the cost of false positives and false negatives are different.
    



"""



```

```python
#  What is overfitting and how can it be prevented?

"""
What is Overfitting?

Overfitting occurs when a machine learning model captures not only the underlying patterns in the training data but also the noise and random fluctuations. As a result, the model performs exceptionally well on the training data but poorly on unseen test data. This happens because the model becomes too complex, tailoring itself too closely to the specific details of the training data rather than generalizing from it.

Signs of Overfitting:
    High Accuracy on Training Data, Low Accuracy on Test Data: The model shows high performance metrics on the training set but significantly worse on the test set.
    
    Complexity: The model uses a high number of parameters or very complex structures (e.g., deep trees in decision trees, high-degree polynomials).
    
    Noise Learning: The model captures random noise in the training data as if it were a true pattern.

How to Prevent Overfitting

Cross-Validation:
    Use techniques like k-fold cross-validation to ensure the model's performance is consistent across different subsets of the data.

Simpler Models:
    Use less complex models that have fewer parameters (e.g., linear regression instead of polynomial regression, shallow decision trees instead of deep ones).

Regularization:
    Apply regularization techniques that add a penalty for larger coefficients to the loss function:
        L1 Regularization (Lasso): Adds the absolute value of coefficients as a penalty term.
        L2 Regularization (Ridge): Adds the square of the coefficients as a penalty term.
        Elastic Net: Combines both L1 and L2 regularization.

Pruning:

    For decision trees, use pruning techniques to remove branches that have little importance and reduce the model's complexity.

Early Stopping:
    In iterative algorithms like gradient descent, stop training when performance on a validation set starts to degrade, rather than continuing to minimize the training error.

Dropout:
    In neural networks, use dropout regularization, which randomly drops a fraction of the neurons during training to prevent over-reliance on particular paths.

Data Augmentation:
    Increase the amount of training data by adding slightly modified copies of existing data or newly created synthetic data to improve the model’s ability to generalize.

Ensemble Methods:
    Combine the predictions of multiple models to reduce overfitting (e.g., bagging, boosting, stacking).

Feature Selection:
    Use techniques to select only the most important features for the model to focus on, reducing the risk of capturing noise from irrelevant features.

Increasing Training Data:
    More data can help the model learn more general patterns rather than specific details. However, this is often difficult to achieve in practice.
    

"""


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

# Generate synthetic data
np.random.seed(0)
X = np.sort(np.random.rand(100) * 10).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Train Ridge Regression model (L2 regularization)
model = Ridge(alpha=1.0)
model.fit(X_train, y_train)

# Predict and plot results
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.plot(X_train, y_pred_train, color='red', label='Model Prediction (Training)')
plt.plot(X_test, y_pred_test, color='orange', label='Model Prediction (Test)')
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.title('Ridge Regression to Prevent Overfitting')
plt.show()
```

```python
#  What is the difference between a classification and a regression problem

"""Difference Between Classification and Regression Problems
Classification and regression are two fundamental types of supervised learning problems in machine learning. They differ primarily in the type of output they predict and the nature of the task they address.

1. Output Type
    Classification:
        The output is categorical.
        The model predicts discrete classes or labels.
        Example: Predicting whether an email is "spam" or "not spam," or determining if a tumor is "benign" or "malignant."
    Regression:
        The output is continuous.
        The model predicts a numerical value.
        Example: Predicting house prices, temperature, or stock prices.
2. Nature of the Task
    Classification:
        The task involves assigning input data to one of several predefined categories.
        Binary Classification: Two classes (e.g., "yes" or "no").
        Multiclass Classification: More than two classes (e.g., "cat," "dog," "rabbit").
    Regression:
        The task involves predicting a real-valued number based on input data.
        The output is often a prediction of a quantity, and the goal is to minimize the error between the predicted and actual values.
3. Evaluation Metrics
    Classification:
        Accuracy: Proportion of correctly predicted labels.
            Precision, Recall, and F1-Score: Especially important in imbalanced datasets.
            AUC-ROC: Used for binary classification to measure the trade-off between true positive rate and false positive rate.
    Regression:
        Mean Absolute Error (MAE): Average absolute difference between predicted and actual values.
        Mean Squared Error (MSE) or Root Mean Squared Error (RMSE): Average squared difference between predicted and actual values.
        R-squared: Proportion of the variance in the dependent variable that is predictable from the independent variables.
4. Example Algorithms
    Classification:
        Logistic Regression, Decision Trees, Random Forests, Support Vector Machines, k-Nearest Neighbors, Neural Networks (with softmax output).
    Regression:
        Linear Regression, Ridge/Lasso Regression, Polynomial Regression, Support Vector Regressor (SVR), Decision Trees, Random Forests (for regression)
        
"""




```

```python
#  Explain the concept of ensemble learning
"""

Ensemble learning is a machine learning technique where multiple models (often referred to as "learners" or "base models") are trained and combined to solve a particular problem. The idea is that by combining the predictions of several models, the overall performance can be improved compared to any individual model. Ensemble methods are particularly effective because different models may capture different patterns or aspects of the data, and their combination can lead to a more robust and accurate prediction.

Key Concepts in Ensemble Learning
   
    Diversity:
    The success of an ensemble method largely depends on the diversity of the base models. If all models make the same errors, combining them won’t help much. However, if different models make different errors, the ensemble can correct those errors by averaging or voting.
    
    Bagging:
        Bootstrap Aggregating (Bagging): A technique where multiple instances of a model are trained on different subsets of the training data, each created by sampling with replacement. The most famous example of bagging is the Random Forest algorithm, where multiple decision trees are trained, and their outputs are averaged for regression or voted on for classification.
        
        Key Benefit: Reduces variance and helps prevent overfitting.
    
    Boosting:
        Boosting: An iterative technique where models are trained sequentially, each trying to correct the errors of the previous one. The final model is a weighted sum of all the models, where models with better performance have more influence.
        Examples: AdaBoost, Gradient Boosting Machines (GBM), XGBoost, LightGBM.
        Key Benefit: Reduces bias and improves the model's predictive accuracy.
    
    Stacking:
        Stacking (Stacked Generalization): An ensemble method where multiple base models are trained on the data, and their predictions are then used as input to a higher-level model (meta-model). The meta-model learns to predict the final output based on the predictions of the base models.
        Key Benefit: Can leverage the strengths of different models and learn a better final prediction.
    
    Voting:
        Voting: A simple ensemble method where the predictions of multiple models are combined by voting. For classification problems, this can be majority voting (the class that gets the most votes wins). For regression problems, predictions can be averaged.
        Key Benefit: Simple and often effective, especially when the base models are diverse.

Examples of Ensemble Methods
    Random Forest (Bagging):

        Trains multiple decision trees on different parts of the data (with replacement) and averages their predictions to improve accuracy and robustness.
    
    AdaBoost (Boosting):

        Sequentially trains weak learners (usually decision stumps) where each new model focuses more on the mistakes made by previous models. The final prediction is a weighted sum of all models.
    
    XGBoost (Extreme Gradient Boosting):

        An optimized and efficient implementation of gradient boosting, widely used in machine learning competitions for its performance and scalability.
    
    Voting Classifier (Voting):

        Combines the predictions of multiple classifiers (e.g., logistic regression, decision tree, SVM) using majority voting for classification tasks.


"""

from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load data
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize base models
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = SVC(probability=True)

# Initialize Voting Classifier
ensemble = VotingClassifier(estimators=[('lr', model1), ('dt', model2), ('svc', model3)], voting='soft')
ensemble.fit(X_train, y_train)

# Predict and evaluate
y_pred = ensemble.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
#  What is gradient descent and how does it work?

"""
Gradient Descent is an optimization algorithm used in machine learning and deep learning to minimize the loss function by iteratively adjusting the model parameters. The goal is to find the optimal set of parameters that result in the lowest possible error between the predicted and actual values. Gradient Descent is widely used for training models such as linear regression, logistic regression, and neural networks.

How Gradient Descent Works
    
    Initialization:

        Start with an initial guess for the model parameters (weights). These can be set randomly or using some heuristic.
    
    Compute the Loss Function:

        The loss function (also called the cost function) measures how far the model's predictions are from the actual target values. Common loss functions include Mean Squared Error (MSE) for regression and Cross-Entropy Loss for classification.
    
    Calculate the Gradient:

        The gradient of the loss function with respect to the model parameters is computed. The gradient is a vector of partial derivatives that indicates the direction in which the loss function increases the most.
    
    Update the Parameters:

        The model parameters are updated by moving them in the opposite direction of the gradient. This step is controlled by a hyperparameter called the learning rate (ɑ), which determines the size of the step taken in each iteration.
        
    Iterate Until Convergence:

        Steps 2-4 are repeated until the algorithm converges to a minimum, or a pre-defined number of iterations (epochs) is reached.
                
"""





```

```python
#  What is the curse of dimensionality in machine learning



"""
The "Curse of Dimensionality" refers to various challenges and problems that arise when analyzing and organizing data in high-dimensional spaces. As the number of features (dimensions) in a dataset increases, the volume of the space increases exponentially, making the data sparse. This sparsity is problematic for machine learning algorithms, which often rely on the density and proximity of data points to make accurate predictions.

Key Issues Associated with the Curse of Dimensionality
    Data Sparsity:

        In high-dimensional spaces, data points become sparse, meaning that the distance between any two points increases. As a result, data points are less likely to be close to each other, making it difficult to find meaningful patterns.
        Many machine learning algorithms, such as k-nearest neighbors (KNN) or clustering algorithms, rely on the concept of distance. In high dimensions, the notion of distance becomes less meaningful because almost all points are far apart.
    
    Increased Computational Complexity:
        As the number of dimensions increases, the computational cost of processing and analyzing the data also increases. Algorithms may take much longer to run or may require more memory, making them less practical for high-dimensional data.
  
    Overfitting:
        High-dimensional datasets are more prone to overfitting because the model can become too complex, capturing noise rather than the underlying pattern. With a large number of features, the model may find spurious correlations that do not generalize to new data.
        Overfitting leads to poor performance on unseen data, as the model becomes too tailored to the training set.
    
    Requirement of More Data:
        As the number of features increases, the amount of data needed to ensure that the model accurately captures the relationships among the variables also increases. Without sufficient data, the model may fail to generalize well.

Example to Illustrate the Curse of Dimensionality

Consider a simple scenario where you want to cluster data points based on their similarity. In a low-dimensional space (e.g., 2D or 3D), the points might be close to each other, making it easy to form clusters. However, as you increase the number of dimensions (e.g., to 100 or 1000 dimensions), the points become spread out, and the distances between them become more uniform. This uniformity makes it difficult to distinguish between close and distant points, complicating the clustering process.

Mitigating the Curse of Dimensionality

    Dimensionality Reduction:

        Principal Component Analysis (PCA): A technique that reduces the dimensionality of the data by projecting it onto the directions (principal components) that capture the most variance.
        t-SNE and UMAP: Non-linear dimensionality reduction techniques that are often used for visualization in lower dimensions.
        Feature Selection: Selecting the most relevant features based on statistical methods, domain knowledge, or model importance metrics (e.g., using Lasso regression).

    Regularization:
        Adding regularization terms (like L1 or L2 regularization) to the model helps prevent overfitting by penalizing overly complex models that rely on too many features.
    
    Collecting More Data:
        Increasing the amount of data can help mitigate the sparsity problem by providing more examples for the model to learn from, improving its ability to generalize.

    Using Specialized Algorithms:
        Some machine learning algorithms are designed to perform better in high-dimensional spaces. For example, tree-based methods like Random Forests and Gradient Boosting Machines are generally less affected by high dimensionality.

    Domain-Specific Techniques:
        Applying domain-specific knowledge to engineer features that are more meaningful and reduce the effective dimensionality.


"""
```
