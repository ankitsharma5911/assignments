10.  Write a code that generates prime numbers up to n using a generator


```python
def is_prime(num):
  """
  This function checks if a number is prime.
  Args:
      num: The number to check.
  Returns:
      True if the number is prime, False otherwise.
  """
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

def prime_generator(n):
  """
  This function generates prime numbers up to n.
  Args:
      n: The upper limit for prime numbers.
  Yields:
      Prime numbers less than or equal to n.
  """
  num = 2
  while num <= n:
    if is_prime(num):
      yield num
    num += 1

# Get prime numbers up to 100
for prime in prime_generator(100):
  print(prime)
```

39. How are abstract methods different from regular methods in Python


1. Abstract Methods:
Abstract methods are declared but not defined within a class.
They serve as a blueprint or interface for other classes to follow.
An abstract class cannot be instantiated directly; it is intended to be inherited by other classes.

2. Regular Methods:
Regular methods (also called concrete methods) are fully defined within a class.
They provide implementation details and can be called directly.
Regular methods can be inherited from a superclass without requiring explicit subclass implementation.

40. How can you achieve abstraction using interfaces in Python


```python
import abc
# 1. define the interface
class Shape(abc.ABC):
    @abc.abstractmethod
    def calculate_area(self):
        pass

# 2. implement the interface
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def calculate_area(self):
        return self.side * self.side

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14159 * self.radius * self.radius
    


def calculate_total_area(shapes):
    total_area = 0
    for shape in shapes:
        total_area += shape.calculate_area()
    return total_area

shapes = [Square(5), Circle(4)]
total_area = calculate_total_area(shapes)
print(total_area) 
```

41. Can you provide an example of how abstraction can be utilized to create a common interface for a group 
of related classes in Python

```python
# 1. Define the Interface (Abstract Base Class):

import abc

class PaymentMethod(abc.ABC):
    @abstractmethod
    def make_payment(self, amount):
        pass

# 2.Implement the Interface (Concrete Subclasses):

class CreditCard(PaymentMethod):
    def __init__(self, card_number, cvv):
        self.card_number = card_number
        self.cvv = cvv

    def make_payment(self, amount):
        # Simulate processing a credit card payment
        print(f"Processing credit card payment of ₹{amount} using number {self.card_number[-4:]}")

class DebitCard(PaymentMethod):
    def __init__(self, card_number, pin):
        self.card_number = card_number
        self.pin = pin

    def make_payment(self, amount):
        # Simulate processing a debit card payment
        print(f"Processing debit card payment of ₹{amount} using number {self.card_number[-4:]}")

class CashOnDelivery(PaymentMethod):
    def make_payment(self, amount):
        # Simulate processing a cash on delivery payment
        print(f"Cash on delivery payment of ₹{amount} will be collected upon delivery")

def process_payment(payment_method, amount):
   payment_method.make_payment(amount)

# Create instances of payment methods
credit_card = CreditCard("1234567890123456", 123)
debit_card = DebitCard("9876543210987654", 4321)
cod = CashOnDelivery()

# Use the common interface for payment processing
process_payment(credit_card, 100)  
process_payment(debit_card, 50)   
process_payment(cod, 25)          

```

45. How does polymorphism improve code readability and reusability


Improved Readability:
Focus on Functionality: Polymorphism allows you to write code that focuses on the "what" (functionality) rather than the "how" (specific implementation details). You define an interface (a set of methods) that different classes can implement in their own ways. This makes the code easier to understand because you're dealing with high-level concepts rather than getting bogged down in the intricacies of each implementation.

Conciseness and Clarity: By using a common interface, you can avoid code duplication. Instead of writing the same logic for different object types, you write it once for the interface and let the subclasses handle the specifics. This leads to shorter, more concise code that's easier to follow and maintain.
Enhanced Reusability:
Generic Code: Polymorphic code is often more generic, meaning it can work with various objects as long as they implement the required interface. This allows you to write code that can be reused in different contexts without modification. For example, a function that calculates the area of a shape can work with squares, circles, rectangles, etc., as long as they all have a method to calculate their area.
Extensibility: Polymorphism makes your code more extensible. You can easily add new classes that implement the interface without having to change the existing code. This promotes code flexibility and adaptability to new requirements.

53.  Modify the decorator to accept arguments and print the function name along with the message


```python
def timing(func):
  """Decorator that prints the execution time of a function."""
  import time

  def wrapper(*args, **kwargs):
    start_time = time.time()
    result = func(*args, **kwargs)
    end_time = time.time()
    print(f"Function '{func.__name__}' executed in {(end_time - start_time):.4f} seconds")
    return result
  return wrapper

def logging(message):
  """Decorator that logs a message before and after function execution."""

  def decorator(func):
    def wrapper(*args, **kwargs):
      print(f"{message} - Executing function: {func.__name__}")
      result = func(*args, **kwargs)
      print(f"{message} - Finished execution of function: {func.__name__}")
      return result
    return wrapper
  return decorator

@logging("DEBUG")  # Applied first
@timing  # Applied second
def my_function(a, b):
  """A simple function that adds two numbers."""
  return a + b

# Call the decorated function
result = my_function(5, 3)
print(f"Result: {result}")
```

56.  Create a decorator that preserves the metadata of the original function


```python
from functools import wraps

def preserve_metadata(func):
  """Decorator that preserves the original function's metadata (docstring, name, etc.)."""

  @wraps(func)
  def wrapper(*args, **kwargs):
    # Call the original function
    result = func(*args, **kwargs)
    return result
  return wrapper

@preserve_metadata
def my_function(a, b):
  """A simple function that adds two numbers."""
  return a + b

# Check if metadata is preserved (using help())
print(help(my_function))
```

65.  What is the purpose of the `__getitem__()` method in Python? Provide an example


```python
class StringList:
  """
  A simple class to represent a list of strings.
  """
  def __init__(self, strings):
    self.strings = list(strings)  # Create a copy of the input list

  # Define the __getitem__() method for accessing elements by index
  def __getitem__(self, index):
    if isinstance(index, int):
      if 0 <= index < len(self.strings):
        return self.strings[index]
      else:
        raise IndexError("Index out of range")
    else:
      raise TypeError("Invalid index type")

  # Define the __str__() method for a readable representation
  def __str__(self):
    return f"StringList: {self.strings}"

# Create a StringList object
string_list = StringList(["apple", "banana", "cherry"])

# Access elements using indexing
first_fruit = string_list[0]
second_fruit = string_list[1]

# Accessing an invalid index (optional)
#try:
#  invalid_fruit = string_list[4]  # This will raise an IndexError
#except IndexError as e:
#  print(f"Error: {e}")

print(f"First fruit: {first_fruit}")
print(f"Second fruit: {second_fruit}")
print(string_list)
```

66. Explain the usage of the `__iter__()` and `__next__()` methods in Python. Provide an example using  iterator


```python
# iter and next are act for loop
class NumberList:
  """
  A simple class that iterates over a sequence of numbers.
  """
  def __init__(self, start, end):
    self.start = start
    self.end = end
    self.current = start - 1  # Initialize current to be before the start

  def __iter__(self):
    return self  # Return itself as the iterator object

  def __next__(self):
    self.current += 1
    if self.current >= self.end:
      raise StopIteration
    return self.current

# Create a NumberList object
number_list = NumberList(1, 5)

# Iterate using a for loop
for number in number_list:
  print(number)
```

71. How does encapsulation relate to property decorators in Python? Provide an example showcasing 
encapsulation using property decorators.

```python
class Book:
  """
  A class representing a book with title and year attributes (encapsulated).
  """
  def __init__(self, title, year):
    self._title = title  # Private attribute
    self._year = year    # Private attribute

  @property
  def title(self):
    """Getter property for the book title."""
    return self._title.upper()  # Optional: formatting applied here

  @title.setter
  def title(self, new_title):
    """Setter property for the book title with validation."""
    if len(new_title) < 2:
      raise ValueError("Book title must be at least 2 characters long")
    self._title = new_title

# Create a Book object
book = Book("The Hitchhiker's Guide", 1979)

# Accessing the title (getter property is called)
print(f"Book title: {book.title}")  # Output: THE HITCHHIKER'S GUIDE (uppercase applied)

# Attempting to set an invalid title (setter validation triggers)
try:
  book.title = "A"
except ValueError as e:
  print(f"Error setting title: {e}")

# Setting a valid title (setter allows it)
book.title = "The Martian"

print(f"Updated book title: {book.title}")  # Output: THE MARTIAN
```
